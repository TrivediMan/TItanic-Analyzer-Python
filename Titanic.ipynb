```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
DF = pd.read_csv('titanic (1).csv')
DF
```

```python
DF.info()
```

```python
DF.head(15)
```

```python
DF.isnull().sum()
```

```python
print("\nOriginal DataFrame (multiple columns):")
print(DF)

DF.drop(['SibSp', 'Parch','Fare','Cabin'], axis=1, inplace=True)

print(DF)
```

```python
Survived_counts = DF['Survived'].value_counts()

plt.figure(figsize=(20,9))

plt.pie(Survived_counts, labels=Survived_counts.index, autopct='%1.1f%%', startangle=120, shadow=True,explode = [0.1, 0],colors = ["Red", "Green"])

plt.title("Survived and Dead")

plt.show()
```

```python
Survived_0 = DF.head(15).groupby('Survived')['Age']
print("Survived by Age:\n", Survived_0, "\n")
```

```python
plt.figure(figsize=(20,9))

Survived_0.plot(kind='line', marker='o', linestyle='-')

plt.title("Survived and Dead")

plt.xlabel("Survived")
plt.ylabel("Age")

plt.legend()

plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.boxplot(x='Sex', y='Age', data=DF, palette="Set3")
plt.title("Box Plot Of Age and Sex")
plt.show()
```

```python
plt.figure(figsize=(15,10))

numeric_df = DF.select_dtypes(include=['number'])
correlation_matrix = numeric_df.corr()

sns.heatmap(correlation_matrix, annot=True, cmap='YlGnBu', fmt=".2f")
plt.title("Correlation Heatmap of Numeric Features")
plt.show()
```

```python
sns.histplot(DF['Age'].dropna(), kde=True, bins=20)
plt.title('Age Distribution of Passengers')
plt.show()
```

```python
plt.figure(figsize=(20,15))

sns.barplot(x = DF["Pclass"], y = DF["Survived"],)

plt.show()
```

```python
DF.groupby(['Sex','Survived'])['Survived'].count()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

f, ax = plt.subplots(1, 2, figsize=(18, 8))

DF[['Sex', 'Survived']].groupby(['Sex']).mean().plot.bar(
    ax=ax[0], 
    color=['skyblue'],
    edgecolor='black'
)
ax[0].set_title('Survived vs Sex')

sns.countplot(
    x='Sex',
    hue='Survived',
    data=DF,
    ax=ax[1],
    palette={0: "red", 1: "green"}
)
ax[1].set_title('Sex: Survived vs Dead')

plt.show()
```
